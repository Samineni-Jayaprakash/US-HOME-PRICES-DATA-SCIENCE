#IMPORTING REQUIRED LIBRARIES

```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
directory_path="cleaned data"
if not os.path.exists(directory_path):
    os.makedirs(directory_path)
```

```python
path = 'cleaned data'

csv_files = [os.path.join(path, f) for f in os.listdir(path) if f.endswith('.csv')]

dfs = [pd.read_csv(f) for f in csv_files]

# Merging the dataframes on the 'DATE' column
final_data= pd.concat(dfs, ignore_index=False).groupby('DATE').sum()
```

## we have unusual points(outliers) errors in features
   1.US HOME PRICE INDEX
   
   2.MEDIAN SALES PRICE
   
   3.UNEMP RATE

****SKEWNESS OF OUTLIERS***

US HOME PRICE INDEX

```python
final_data["US HOME PRICE INDEX"].skew()
```

```python
q1=final_data["US HOME PRICE INDEX"].quantile(0.25)
q2=final_data["US HOME PRICE INDEX"].quantile(0.75)
iqr=q2-q1
```

```python
lower_bridge=q1-(iqr*1.5)
upper_bridge=q2+(iqr*1.5)
```

```python
lower_bridge
```

```python
upper_bridge
```

```python
iqr
```

```python
final_data.loc[final_data['US HOME PRICE INDEX']>=upper_bridge,'US HOME PRICE INDEX']=upper_bridge
```

```python
final_data['US HOME PRICE INDEX'].hist()
plt.title("distribution of us home price index after handling outliers")
plt.xlabel("US HOME PRICE INDEX")
plt.ylabel('count')
plt.show()
```

```python
final_data["US HOME PRICE INDEX"]=np.log1p(final_data["US HOME PRICE INDEX"])
```

```python
plt.boxplot(final_data["MEDIAN SALES PRICE"],vert=False)
plt.xlabel("MEDIAN SALES PRICE")
plt.ylabel('boxplot of median sales price')
plt.show()
```

### Checking Multicollinearity

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

var_inf_fact_data = pd.DataFrame()
var_inf_fact_data["feature"] = final_data.columns

var_inf_fact_data["VIF"] = [variance_inflation_factor(final_data.values, i)
                          for i in range(len(final_data.columns))]

var_inf_fact_data
```

##### we have multi collinearity in our data

```python
var_inf_fact_data.plot()
```
